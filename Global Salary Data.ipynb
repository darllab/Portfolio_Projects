### Group By and Aggregating

```python
group_by_frame = df.groupby('continent_name')
```

```python
# By using group by function, it displays median, average, lowest and highest salary in each continent.

group_by_frame.mean()
```

```python
df.groupby('continent_name').mean()
```

```python
# In the groupby function Count, this is just a count of countries per continent

df.groupby('continent_name').count()
```

```python
# Displaying the sum of salary of each continent

df.groupby('continent_name').sum()
```

```python
# This is a better view to compare which continent has lowest and highest salary 

df.groupby('continent_name').agg({'lowest_salary': ['mean','max','count','sum'], 'highest_salary':['mean','max','count','sum'] })
```
